```python
from sec_api import QueryApi
```

```python
API_KEY = ''
ticker = 'DRI'

query_api = QueryApi(api_key=API_KEY)
```

```python
query = {
  "query": "ticker:DRI AND filedAt:[2018-01-01 TO 2024-03-31] AND formType:\"10-Q\"",
  "from": "0",
  "size": "30",
  "sort": [{ "filedAt": { "order": "desc" } }]
}

filings = query_api.get_filings(query)
```

```python
import pickle

with open('DRI_10Q_filing.pkl', 'wb') as f:
    pickle.dump(filings, f, protocol=pickle.HIGHEST_PROTOCOL)
```
